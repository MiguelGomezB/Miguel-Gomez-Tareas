```python
import numpy as np
import matplotlib.pyplot as plt
```

#Ejercicio A2

```python

def f1(n,epsilon):
    x = np.linspace(0,1,n+1)
    funcion_1 = np.sin(1/(x + epsilon))
    plt.plot(x, funcion_1,c="blue", marker='o', label=f'n={n}, ε={epsilon}')
    plt.xlabel('x')
    plt.ylabel(r'$\sin\left(\frac{1}{x + \epsilon}\right)$')
    plt.title('Gráfico de f(x) = sin(1/(x+ε))')
    plt.legend()
    plt.grid(True)
    return(plt.show())
```

```python
f1(10,1/5)
```

```python
def f1(n,m,epsilon):
    x_1 = np.linspace(0,1,n+1)
    x_2 = np.linspace(0,1,n+m+1)
    funcion_1 = np.sin(1/(x_1 + epsilon))
    funcion_2 = np.sin(1/(x_2 + epsilon))
    plt.plot(x_1, funcion_1,c="blue", marker='o', label=f'n={n}, ε={epsilon}')
    plt.plot(x_2, funcion_2,c="green", marker='o', label=f'n={n+m}, ε={epsilon}')
    plt.xlabel('x')
    plt.ylabel(r'$\sin\left(\frac{1}{x + \epsilon}\right)$')
    plt.title('Gráfico de f(x) = sin(1/(x+ε))')
    plt.legend()
    plt.grid(True)
    return(plt.show())
```

(C) How large do you have to choose n in order for the difference between these two functions to be less than 0.1? Hint: Each function gives an array. Create a while loop and use the max function of the arrays to retrieve the maximum value and compare these

```python
def f1(n,m,epsilon):
    x_1 = np.linspace(0,1,n+1)
    x_2 = np.linspace(0,1,n+m+1)
    funcion_1 = np.sin(1/(x_1 + epsilon))
    funcion_2 = np.sin(1/(x_2 + epsilon))
    return np.abs(np.max(funcion_1)-np.max(funcion_2))
```

```python
f1(2,10,1/5)
#Se necesita n = 2 para epsilon = 1/5
```

```python
f1(2,10,1/10)
#Se necesita n = 2 para epsilon = 1/10
```

```python
f1(1,10,1/20)
#Se necesita n = 1 para epsilon = 1/20
```
